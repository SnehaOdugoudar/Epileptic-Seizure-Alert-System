```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report
import os
```

```python
%%time
list_of_pca_numbers = [20, 50, 70, 100]

for i in list_of_pca_numbers:
    print("Running pca - %d..." % (i))
    train_path = 'edgeml/examples/pytorch/ProtoNN/pca_' + str(i) + '_test_20/train.npy'
    test_path = 'edgeml/examples/pytorch/ProtoNN/pca_' + str(i) + '_test_20/test.npy'
    
    neighbors = np.arange(1, 40)
    train_accuracy = np.empty(len(neighbors))
    test_accuracy = np.empty(len(neighbors))
    
    train, test = np.load(train_path), np.load(test_path)
    
    x_train, y_train = train[:, 1:], train[:, 0]
    x_test, y_test = test[:, 1:], test[:, 0]
    
    for j, k in enumerate(neighbors):
        
        knn = KNeighborsClassifier(n_neighbors = k)
        knn.fit(x_train, y_train)
        train_accuracy[j] = knn.score(x_train, y_train)
        test_accuracy[j] = knn.score(x_test, y_test)
        
#     print("Maximum Training accuracy : ", max(train_accuracy), " at K value - ", np.where(train_accuracy == max(train_accuracy)))
    print("Maximum Testing accuracy : ", max(test_accuracy), " at K value - ", np.where(test_accuracy == max(test_accuracy)))
    print("Classification Report : \n", classification_report(knn.predict(x_test), y_test))
    print("*" * 30)
    plt.plot(neighbors, test_accuracy, label = 'Testing dataset Accuracy')
    plt.plot(neighbors, train_accuracy, label = 'Training dataset Accuracy')

    plt.legend()
    plt.xlabel('n_neighbors')
    plt.ylabel('Accuracy')
    plt.show()
```
